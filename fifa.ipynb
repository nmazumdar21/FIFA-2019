```python
def to_Int(data_value):
  
  try:
    value = float(data_value[1:-1])
    sign = data_value[-1:]

    if sign=="M":
      value = value*1000000
    elif sign =="K":
      value = value*1000
  except ValueError:
    value = 1
  return value
```
